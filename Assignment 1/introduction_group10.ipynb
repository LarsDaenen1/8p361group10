### Deep learning for medical image analysis 

In the coming decade, artificial intelligence will affect many aspects of human life and society ranging from home automation and self-driving cars to algorithmic detection of “fake news” on social media. This technology will also inevitably influence healthcare and in turn medical imaging. One of the most active fields of research in artificial intelligence is deep machine learning [1][2]. This concept works by learning abstract, hierarchical feature representations of input data that are predictive of a certain target (e.g. a class label in the case of classification). In contrast to classical machine learning methods, deep learning eliminates the step of “manual” feature engineering that is a tedious process and sub-optimal compared with learned representations.

In the field of medical image analysis, the use of deep learning methods has already resulted in notable progress for a variety of tasks. Some of the first applications where automatic methods were shown to outperform medical experts have been in histopathology image analysis [3]. 


### Medical image analysis challenges

One of the developments that brought about the success of deep learning methods in computer vision was the availability of large, public annotated image datasets. In medical image analysis, this was mirrored in the form of medical image analysis challenges – friendly competitions in which researchers worldwide evaluate their solutions on the same data with the same criteria, in a blinded manner [4]. Organizing a challenge constitutes collecting and publicly distributing a training dataset that can be used to address a specific clinical task (such as detection of lymph node metastases). The ground truth for the testing is set is retained by the organizers of the challenge, which facilitates fair and independent evaluation and comparison of methods. The organization of medical image analysis challenges has had a tremendous impact on the advancement of the state of the art in the field. In the coming decade, such efforts will continue to push the field forward.

Some other examples of histopathology image analysis challenges organized by our group are AMIDA13 [5] and TUPAC16 [6].

### Detection of lymph node metastases

The presence of metastases in sentinel axillary lymph nodes is an important component of breast cancer staging [7]. This is evaluated by examination of histological preparations of lymph node tissue by pathologists. The figure below gives an example of a metastasis in lymph node tissue. Although manual assessment by a pathologists is the current standard practice, this procedure is notoriously tedious to perform and can result in missed metastases, which in turn results in suboptimal therapy planning for the patients. The move towards digitalization of histopathology labs with the introduction of whole-slide image scanners [8], has opened the possibility of automating some of the tasks performed by pathologists with the use of image analysis methods. The goal is to have faster, more accurate and more objective analysis of the tissue. Tedious and error prone tasks such as detection of metastases are a very good use case for introduction of automated image analysis.

## Exercise 1

What is the clinical utility of evaluating the presence of metastases in sentinel lymph nodes in breast cancer patients? In other words, how is this information used in the clinical decision making process for breast cancer patients?

Sentinel lymph nodes (SLNs) are the first lymph nodes “to which cancer cells are most likely to spread from a primary tumor” (Sentinel Lymph Node Biopsy, 2019). Histologic examination of the SLNs can reveal whether metastasis to the SLNs has occurred, and can thus provide an indication of metastasis to more distant sites in the body (Donegan, 1997).  Drainage from the breast is dependent on the site of the tumor, as drainage can occur to both the axillary lymph nodes (ALNs) or the internal mammary lymph nodes (IMLNs) (Estourgie et al., 2004).  
The most indicative predictor for death and posttreatment recurrence in breast cancer is the presence (or absence) of metastasis to the axillary lymph nodes (ALN). The absolute number of involved nodes (e.g. nodes with cancer cells present) can provide a more quantitative answer for recurrence and survival, as the 10-year survival decreases with the number of involved nodes. Not only axillary, but also the IMLNs provide proof for possible metastasis. It has been shown that even when the ALNs are metastasis-free, the IMLNs are involved in nine percent of cases. This indicates that investigation of IMLNs can provide extra information to whether a tumor is locally confined or not (Donegan, 1997).
Evaluating the presence of metastases in SLNs determines whether axillary lymph node dissection (ALND) is needed. When no metastasis or isolated tumor cells are found, the ALNs can be spared and ALND is not performed. When there is a metastasis present (micro- or macro-metastasis), ALND will take place. This is because for larger metastases, the prevalence of ALN involvement significantly increases. For patients with SLN micro-metastasis, ALND is especially important as patients not undergoing ALND have an increased risk of recurrence. Furthermore, lympho-vascular invasion in primary tumors has been shown to correlate significantly with SLN positivity (Illyes et al., 2014).

## Exercise 2 

How does the introduction of whole-slide imaging change the typical workflow of a pathology lab?

Prior to the introduction of whole-slide imaging (WSI), digital imaging in anatomic pathology relied largely on microscope-mounted cameras to produce static digital images. These images were limited in utility because they captured only specific regions of a glass slide. WSI technology was inspired by pioneering efforts to achieve high-resolution scanning of an entire glass slide. Most modern WSI instruments today are capable of producing high-resolution digital slides in the span of minutes (or less) [1]. 
WSI has been used for a wide variety of clinical purposes, including telepathology for primary diagnosis, consultation (second opinions), and remotely interpreting frozen sections. Whole slide images have also been used for remotely viewing immunostains, showcasing pathology slides at tumor boards, saving selected consult slides before they are returned, archiving slides before they get sacrificed to use cellular material for molecular studies, and performing image analysis. [2]
Full digitalization of the microscopic evaluation of stained tissue sections in histopathology has become feasible in recent years because of advances in slide scanning technology and cost reduction in digital storage. Advantages of digital pathology include remote diagnostics, immediate avail- ability of archival cases and easier consultations with expert pathologists. Also, the possibility for computer-aided diagnostics may be advantageous. [3]
Accurate breast cancer staging is an essential task performed by pathologists worldwide to inform clinical management. Assessing the extent of cancer spread by histopathological analysis of sentinel axillary lymph nodes (SLNs) is an important part of breast cancer staging. The sensitivity of SLN assessment by pathologists, however, is not optimal. A retrospective study showed that pathology review by experts changed the nodal status in 24% of patients. Furthermore, SLN assessment is tedious and time consuming. It has been shown that deep learning algorithms could identify metastases in SLN slides with 100% sensitivity, whereas 40% of the slides without metastases could be identified as such. This could result in a significant reduction in the workload of pathologists. [3]

The PatchCamelyon dataset is derived from the CAMELYON16 dataset of whole-slide images. Describe how a neural network classification model trained on small image patches can be applied to larger, whole-slide images with the goal of detecting metastases.


A gigapixel scale whole-slide image (WSI) can not be easily analyzed due to the immeasurable computational cost [2], in such it can be split up in small image patches, which can be used to train a neural network classification model.
Furthermore the differentiation of cancer subtypes is based on cellular-level visual features observed on image patch scale [1]. All features that are necessary for detection, and recognition, of metastases can be found in images the size of patches. 
Division of a WSI into small patches, and subsequent determination of metastase presence for each of these patches is a WSI patch-by-patch analyzing method for metastases. The results of this method can be combined to fully analyze a WSI for metastase presence; If all subdivisions of a WSI show no metastase it can be concluded that the tissue Is healthy. If some subdivisions show metastases according to the model, it can be concluded that metastases are present in the WSI.
A heatmap of metastase prediction level per patch on the WSI could also be generated to aid in diagnosis.

`````python
### Python script that reads and displays a few images from the two classes. ###
import os, random
import matplotlib.pyplot as plt
pad_0 = r"C:\Users\20192735\Documents\8P361 - images\train\0"
pad_1 = r"C:\Users\20192735\Documents\8P361 - images\train\1"
pad = r"C:\Users\20192735\Documents\8P361 - images\train"
filenames = [[], []]
num_samples = 6
while len(filenames[0]) != num_samples or len(filenames[1]) != num_samples:
    if len(filenames[0]) != num_samples:
        file_0 = random.choice(os.listdir(pad_0)) 
        if file_0 not in filenames[0]:
            filenames[0].append(file_0)
    if len(filenames[1]) != num_samples:
        file_1 = random.choice(os.listdir(pad_1))
        if file_1 not in filenames[1]:
            filenames[1].append(file_1)
  
print(filenames)  
fig = plt.figure(figsize=(15,5))
for i in range(num_samples):
    img_0 = plt.imread(pad_0+"\\"+filenames[0][i])
    ax = fig.add_subplot(2, num_samples, i+1, frameon=False)
    ax.axes.get_xaxis().set_visible(False)
    ax.axes.get_yaxis().set_visible(False)
    ax.set_title('0')
    plt.imshow(img_0)
    
    img_1 = plt.imread(pad_1+"\\"+filenames[1][i])
    ax = fig.add_subplot(2, num_samples, i+num_samples+1, frameon=False)
    ax.axes.get_xaxis().set_visible(False)
    ax.axes.get_yaxis().set_visible(False)
    ax.set_title('1')
    plt.imshow(img_1)    
    
    
plt.savefig(r"C:\Users\20192735\OneDrive - TU Eindhoven\8P361 - Project Imaging\8p361-project-imaging-master\assignments\Lymph nodes.png")
plt.show()
`````
